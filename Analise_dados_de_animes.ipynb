# Task
Analyze the dataset "anime.csv" to identify interesting information and insights.

## Explore numerical features

### Subtask:
Analyze the distribution and summary statistics of numerical columns such as `episodes`, `rank`, `popularity`, `members`, `favorites`, `score`, and `scored_by`. Identify any correlations between these features.


**Reasoning**:
Select the numerical columns, calculate and display descriptive statistics, compute the correlation matrix, and display the correlation matrix.



```python
numerical_cols = ['episodes', 'rank', 'popularity', 'members', 'favorites', 'score', 'scored_by']
numerical_df = df[numerical_cols]

display(numerical_df.describe())

correlation_matrix = numerical_df.corr()
display(correlation_matrix)
```

## Explore categorical features

### Subtask:
Analyze the distribution of categorical columns such as `status`, `rating`, `genres`, `themes`, and `demographics`. Identify the most frequent categories in each column.


**Reasoning**:
Select the specified categorical columns and calculate the value counts for each to analyze their distributions.



```python
categorical_cols = ['status', 'rating', 'genres', 'themes', 'demographics']

for col in categorical_cols:
    print(f"Value counts for column: {col}")
    display(df[col].value_counts())
    print("-" * 30)
```

## Analyze relationships between numerical and categorical features

### Subtask:
Explore how categorical features influence numerical features (e.g., how does the rating affect the score or popularity?).


**Reasoning**:
Select the specified categorical and numerical columns, then group by each categorical column and calculate the mean of the specified numerical columns.



```python
categorical_cols = ['status', 'rating', 'genres', 'themes', 'demographics']
numerical_cols = ['score', 'popularity']

for col in categorical_cols:
    if col in df.columns:
        print(f"Mean score and popularity by {col}:")
        grouped_df = df.groupby(col)[numerical_cols].mean()
        display(grouped_df)
        print("-" * 30)
    else:
        print(f"Column '{col}' not found in the DataFrame.")
```

## Analyze time-series data

### Subtask:
Convert `aired_from` and `aired_to` to datetime objects and analyze trends over time, such as the number of anime released per year or the average score over time.


**Reasoning**:
Convert the 'aired_from' and 'aired_to' columns to datetime objects, extract the year from 'aired_from', calculate the number of anime released per year, and calculate the average score per year.



```python
df['aired_from'] = pd.to_datetime(df['aired_from'], errors='coerce')
df['aired_to'] = pd.to_datetime(df['aired_to'], errors='coerce')

df['release_year'] = df['aired_from'].dt.year

anime_per_year = df['release_year'].value_counts().sort_index()
average_score_per_year = df.groupby('release_year')['score'].mean()

print("Number of anime released per year:")
display(anime_per_year)

print("\nAverage score per year:")
display(average_score_per_year)
```

## Identify missing values

### Subtask:
Analyze the missing values in the dataset and consider strategies for handling them (e.g., imputation, removal).


**Reasoning**:
Calculate and display the number and percentage of missing values for each column to understand the extent of missing data.



```python
missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100

print("Number of missing values per column:")
display(missing_values)

print("\nPercentage of missing values per column:")
display(missing_percentage)
```

## Analyze text data

### Subtask:
Explore the `title`, `title_english`, `title_japanese`, and `synopsis` columns using text analysis techniques to find common keywords or themes.


**Reasoning**:
Select the text columns, handle missing values, combine the text, apply basic preprocessing, and use TF-IDF to vectorize the text for keyword extraction.



```python
from sklearn.feature_extraction.text import TfidfVectorizer
import re

text_cols = ['title', 'title_english', 'title_japanese', 'synopsis']

df[text_cols] = df[text_cols].fillna('')

df['combined_text'] = df[text_cols].agg(' '.join, axis=1)

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    return text

df['preprocessed_text'] = df['combined_text'].apply(preprocess_text)

tfidf = TfidfVectorizer(max_features=1000, stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['preprocessed_text'])

feature_names = tfidf.get_feature_names_out()
```

**Reasoning**:
Analyze the TF-IDF matrix to find the most frequent terms or patterns by examining the terms with the highest TF-IDF scores across the dataset.



```python
import numpy as np

average_tfidf_scores = tfidf_matrix.mean(axis=0).A1
tfidf_ranking = np.argsort(average_tfidf_scores)[::-1]

print("Top 20 most frequent terms based on average TF-IDF score:")
for i in range(20):
    print(f"{feature_names[tfidf_ranking[i]]}: {average_tfidf_scores[tfidf_ranking[i]]:.4f}")
```
