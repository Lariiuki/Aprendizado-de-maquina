**Aprimoramento da coluna 'episodes'**

Observamos que a coluna 'episodes' contém valores inconsistentes, especialmente para animes de longa duração, que possuem número de episódios muito maior do que o registrado ou até mesmo valores zerados. Para lidar com essas inconsistências, faremos um tratamento dos dados convertendo a coluna em uma variável categórica, facilitando a análise e a interpretação dos diferentes perfis de duração dos animes.

```python
import numpy as np
df_anime_filtered_demographics['episodes'] = df_anime_filtered_demographics['episodes'].replace(0, np.nan)
```

```python
df_known_episodes = df_anime_filtered_demographics[df_anime_filtered_demographics['episodes'] > 0]
```

```python
def categorize_episodes(x):
    if pd.isna(x):
        return "Desconhecido"
    elif x == 0:
        return "Iniciando (>5)"
    elif x < 12:
        return "Curto (<12)"
    elif x < 50:
        return "Médio (12-49)"
    else:
        return "Longo (50+)"

df_anime_filtered_demographics['episodes_category'] = df_anime_filtered_demographics['episodes'].apply(categorize_episodes)
```

**Associação entre variáveis:** Quais pares numéricos têm maior correlação?

```python
num_cols = df_anime_filtered_demographics.select_dtypes(include='number').columns.tolist()
corr = df_anime_filtered_demographics[num_cols].corr()

plt.figure(figsize=(6,5))
plt.imshow(corr, interpolation='nearest')
plt.title("Matriz de Correlação")
plt.colorbar()
plt.xticks(ticks=range(len(num_cols)), labels=num_cols, rotation=90)
plt.yticks(ticks=range(len(num_cols)), labels=num_cols)
plt.tight_layout()
plt.show()

pairs = []
for i in range(len(num_cols)):
    for j in range(i+1, len(num_cols)):
        pairs.append((num_cols[i], num_cols[j], corr.iloc[i, j]))
pairs_sorted = sorted(pairs, key=lambda x: abs(x[2]), reverse=True)[:3]
print("Resposta P3: Maiores correlações (valor absoluto):")
for a,b,v in pairs_sorted:
    print(f" - {a} x {b}: {v:.2f}")
```

```python
numeric_cols = df_anime_filtered_demographics.select_dtypes(include='number')

plt.figure(figsize=(10, 8))
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de Correlação')
plt.show()
```
