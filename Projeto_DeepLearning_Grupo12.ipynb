# Data Cleasing - Formatação do dataset

**Formatando as datas**

Percebemos que as datas estão no formato data e hora e não utilizaremos a hora para nenhuma análise. Iremos retirar e deixarmos no formato somente de data.

```python
# Converter 'aired_from' e 'aired_to' para datetime, mantendo apenas a data
df_anime['aired_from'] = pd.to_datetime(df_anime['aired_from'], errors='coerce')
df_anime['aired_to'] = pd.to_datetime(df_anime['aired_to'], errors='coerce')
```

```python
# Exibir as primeiras linhas para verificar a transformação
df_anime[['aired_from', 'aired_to']].head()
```

```python
# Encontrar a data mínima e máxima nas colunas 'aired_from' e 'aired_to'
min_aired_from = df_anime['aired_from'].min().strftime('%d/%m/%Y')
max_aired_from = df_anime['aired_from'].max().strftime('%d/%m/%Y')
min_aired_to = df_anime['aired_to'].min().strftime('%d/%m/%Y')
max_aired_to = df_anime['aired_to'].max().strftime('%d/%m/%Y')

print(f"Range da data de início de exibição (aired_from): {min_aired_from} a {max_aired_from}")
print(f"Range da data de fim de exibição (aired_to): {min_aired_to} a {max_aired_to}")
```

**Formatação de gêneros**

```python
df_anime.head()
```

```python
# Verificar os valores únicos na coluna 'genres' e sua contagem
print("Contagem de valores na coluna 'genres':")
display(df_anime['genres'].value_counts())

# Opcional: Ver os valores únicos (sem contagem)
#print("\nValores únicos na coluna 'genres':")
#display(df['genres'].unique())
```

```python
# Lidar com valores ausentes na coluna 'genres' preenchendo com uma string vazia para evitar erros ao splitar
df_anime['genres'] = df_anime['genres'].fillna('')

# Separar os gêneros em listas
df_anime['genres'] = df_anime['genres'].str.split(', ')

# Exibir as primeiras linhas para verificar a nova coluna
print("DataFrame com a coluna 'genres' como lista:")
display(df_anime['genres'].head())
```

```python
# Se quiser manter a coluna com os gêneros filtrados e sem strings vazias:
df_anime['genres'] = df_anime['genres'].apply(lambda gs: [g for g in gs if g != ''])
```

```python
df_anime.head()
```

**Formatação dos temas**

```python
# Lidar com valores ausentes na coluna 'genres' preenchendo com uma string vazia para evitar erros ao splitar
df_anime['themes'] = df_anime['themes'].fillna('')

# Separar os gêneros em listas
df_anime['themes'] = df_anime['themes'].str.split(', ')

# Exibir as primeiras linhas para verificar a nova coluna
print("DataFrame com a coluna 'genres' como lista:")
display(df_anime['themes'].head())
```

```python
# Se quiser manter a coluna com os gêneros filtrados e sem strings vazias:
df_anime['themes'] = df_anime['themes'].apply(lambda gs: [g for g in gs if g != ''])
```

**Observações sobre a qualidade dos dados**

Durante a análise exploratória, foi identificado que algumas colunas possuem um número substancial de valores ausentes. As colunas 'aired_from', 'aired_to', 'release_year', 'title_english', 'demographics', 'themes', 'genres', 'score', 'scored_by' e 'rank' são as que mais demandam estratefias de tratamento de dados faltantes que vai depender dos objetivos da análise.

```python
df_anime[['score', 'scored_by', 'rank']].describe()
```

Olhando para a média (6.39) e a mediana (50% é igual 6.37), elas são bem próximas. Isso sugere que a distribuição da coluna 'score' pode não ter muitos outliers extremos que distorceriam a média significantivamente. No entando, a coluna 'scored_by' (número de usuários que avaliaram), que está relacionada, tem uma grande diferença entre a média e a mediana, indicando outiliers.

**Removendo valores do dataset**

Durante a fase de limpeza e pré-processamento do dataset de animes, identificamos que a coluna 'demographics' apresentava uma quantidade significativa de valores ausentes, correspondendo a aproximadamente 63% do total de entradas. Considerando os objetivos deste projeto, que incluem a análise de características e tendências dos animes, a informação demográfica é um atributo potencialmente relevante para segmentar e compreender o público-alvo de diferentes títulos.

Diante da alta taxa de dados faltantes e da natureza categórica da coluna 'demographics', optamos pela estratégia de remover as linhas que possuíam valores nulos especificamente nesta coluna. Esta decisão foi tomada com base nos seguintes argumentos:



Alternativas Menos Ideais: Outras estratégias, como a imputação pela moda, poderiam super-representar uma única demografia, enquanto a criação de uma categoria 'Desconhecido' poderia complicar as análises comparativas entre as demografias conhecidas. Dado o foco potencial em analisar as demografias específicas presentes no dataset, a remoção se mostrou a abordagem mais direta para isolar e estudar esses grupos.

1.   Foco Analítico: Para realizar análises que dependem da informação demográfica (por exemplo, comparar pontuações, popularidade ou gêneros entre diferentes demografias como Shounen, Shoujo, Seinen, Josei, Kids), é fundamental ter dados completos e confiáveis para essa variável. A imputação de uma coluna categórica com tantos valores ausentes poderia introduzir ruído ou vieses significativos na análise, distorcendo as conclusões sobre as características de cada demografia.
Preservação da Qualidade da Análise
2.   Demográfica: Ao remover as linhas com 'demographics' nulo, garantimos que qualquer análise subsequente que utilize esta coluna será baseada apenas em dados onde a demografia é conhecida e explicitamente definida. Isso aumenta a confiança nos resultados obtidos ao analisar as diferenças ou padrões entre os grupos demográficos.
3. Tamanho Suficiente do Dataset Remanescente: Apesar da remoção resultar na perda de uma parte considerável do dataset original, o número de linhas restantes (aproximadamente 10.788) ainda excede significativamente o requisito mínimo de 1000 linhas estabelecido para o projeto. O volume de dados remanescente é mais do que adequado para realizar as análises planejadas, incluindo aquelas que não dependem da demografia.
4.


Portanto, a remoção das linhas com valores ausentes na coluna 'demographics' foi uma escolha deliberada para garantir a qualidade e a validade das análises que envolvem este atributo crucial, priorizando a integridade da informação demográfica para os animes onde ela está disponível, sem comprometer a viabilidade do projeto em termos de volume de dados

```python
# Remover linhas com valores ausentes na coluna 'demographics'
df_anime_filtered_demographics = df_anime_v2.dropna(subset=['demographics']).copy()

# Verificar o número de linhas restantes
print(f"Número de linhas após remover NaNs em 'demographics': {len(df_anime_filtered_demographics)}")

# Verificar se ainda existem valores ausentes na coluna 'demographics'
print("\nNúmero de valores ausentes na coluna 'demographics' no novo DataFrame:")
display(df_anime_filtered_demographics['demographics'].isnull().sum())

# Exibir as primeiras linhas do novo DataFrame
display(df_anime_filtered_demographics.head())
```

**Quais são os tipos de generos mais populares dos animes?**

```python
import seaborn as sns
from matplotlib import pyplot as plt

# Obter os dados de contagem usando a Series 'all_genres' criada na célula anterior
all_genres = df_anime_filtered_demographics['genres'].explode()
genre_counts = all_genres.value_counts()

# Criar o gráfico de barras
plt.figure(figsize=(14, 7))
ax = sns.barplot(x=genre_counts.index, y=genre_counts.values, palette='viridis')

# Rotacionar os rótulos do eixo x para melhor legibilidade
plt.xticks(rotation=90)

# Adicionar título e rótulos aos eixos
plt.title('Contagem de Animes por Gênero', fontsize=16)
plt.xlabel('Gênero', fontsize=12)
plt.ylabel('Número de Animes', fontsize=12)

# Adicionar os números acima de cada barra
for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', padding=3)

# Exibir o gráfico
plt.tight_layout() # Ajusta o layout para evitar sobreposição
plt.show()
```

# Análise temporal do volume de produção

```python
# análise temporal do volume de produção e a qualidade percebida ao longo do tempo.
df['aired_from'] = pd.to_datetime(df['aired_from'], errors='coerce')
df['aired_to'] = pd.to_datetime(df['aired_to'], errors='coerce')

df['release_year'] = df['aired_from'].dt.year

anime_per_year = df['release_year'].value_counts().sort_index().reset_index()
average_score_per_year = df.groupby('release_year')['score'].mean()

print("Number of anime released per year:")
display(anime_per_year)

print("\nAverage score per year:")
display(average_score_per_year)
```
